```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
cropped_img = cv2.imread('pictures/green-cropped.jpg')
hsv_test_image = cv2.cvtColor(cropped_img, cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(hsv_test_image[...,1])
```

```python
s_val_1d = hsv_test_image[:,:,1].ravel()

plt.hist(s_val_1d,bins=50)
plt.xlabel("saturation")
plt.ylabel("occurences")
plt.show()
```

```python
saturation_threshold=30
idx = hsv_test_image[:, :, 1] >= saturation_threshold
sat_mask = np.ones_like(hsv_test_image[:, :, 1])
sat_mask[idx] = 0
fig,ax=plt.subplots(nrows=1,ncols=2)
ax[0].imshow(hsv_test_image[:,:,1])
ax[1].imshow(sat_mask)
plt.show()
```

```python
val_threshold = 50
idx = hsv_test_image[:, :, 0] < val_threshold

sat_mask = np.zeros_like(hsv_test_image[:, :, 0])

sat_mask[idx] = 0
fig,ax=plt.subplots(nrows=1,ncols=2)
ax[0].imshow(hsv_test_image[:,:,0])
ax[1].imshow(sat_mask)
plt.show()
```
